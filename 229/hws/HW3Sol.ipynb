# Problem 1. [10 x 5 = 50 points] Set Convexity

(a) **Prove or disprove** convexity of the set 
$$\{c\in\mathbb{R}^{n+1} \mid -5 \leq c_0 + c_1\sin(x) + c_2\sin(2x) + ... + c_{n}\sin(n x) \leq 7 \; \text{for}\; 2022 \leq x \leq 2023\}.$$

(b) **Prove or disprove** convexity of the set
$$\bigg\{x\in\mathbb{R}^{n} \mid \frac{|x_1|^{p}}{a_1^p} + \frac{|x_2|^{p}}{a_2^p} + ... + \frac{|x_n|^{p}}{a_n^p} \leq 1\bigg\}, \quad \text{for fixed}\quad p\ge 1, \;a_i > 0 \;\forall\; i=1, ..., n.$$

(c) **Prove** the convexity of the following set
$$\{x\in\mathbb{R}^{n} \mid \|x -a \|_2 \leq k \|x-b\|_2, \; a\neq b, \; 0\leq k \leq 1\}.$$

(d) Consider a discrete (scalar) random variable $x$ having $n$ possible outcomes $a_1,...,a_n$ with respective probabilities $\mathbb{P}\left(x=a_i\right)=p_i$ for all $i=1,...,n$. You already learnt that the probability vector $p:=(p_1,...,p_n)^{\top}\in\Delta^{n-1}$ (standard simplex). Given nonlinear $f:\mathbb{R}\mapsto\mathbb{R}$, we define its expectation $\mathbb{E}f(x) := \sum_{i=1}^{n}p_i f(a_i)$. For any $\alpha  < \beta$, **prove** that the following set is convex:
$$\{p\in\Delta^{n-1}\mid \alpha \leq \mathbb{E}f(x) \leq \beta\}.$$

(e) Consider the same notations as in part (d). Define the variance ${\rm{var}}(x) := \mathbb{E}x^2 - \left(\mathbb{E}x\right)^2$. **Give a simple counter-example** to show nonconvexity of the set
$$\{p\in\Delta^{n-1}\mid {\rm{var}}(x) \leq \alpha\}, \quad \text{for fixed}\quad\alpha >0.$$

# Solution for Prob. 1.

(a) For a fixed $x\in\left[2022,2023\right]$ let $\mathcal{C}_{x} := \{c\in\mathbb{R}^{n+1} \mid -5 \leq c_0 + c_1\sin(x) + c_2\sin(2x) + ... + c_{n}\sin(n x) \leq 7\}$ which is the intersection of two halfspaces corresponding to the two linear inequalities in $c\in\mathbb{R}^{n+1}$, and therefore a convex set (in fact a polyhedron) because halfspaces are convex sets. 

The given set is the (uncountably many) intersection of convex sets: $\bigcap_{2022 \leq x \leq 2023}\mathcal{C}_{x}$, and hence by Lec. 7, p. 3 bottom, a **convex set**.

(b) For fixed $p\geq 1$, consider the unit $p$ norm ball $\mathcal{B}_{p} := \bigg\{x\in\mathbb{R}^{n} \mid |x_1|^{p} + |x_2|^{p} + ... + |x_n|^{p} \leq 1\bigg\}$ which is a convex set (Lec. 6, p. 7-8).

Letting $A:={\rm{diag}}\left(a_1,a_2,...,a_n\right)$, the given set is the linear transform of the unit $p$ norm ball: $A\mathcal{B}_{p}$. By Lec. 7, p. 3, any affine transform of a convex set is convex, and hence the special case of a linear map with diagonal $A$ preserves convexity too. Thus, $A\mathcal{B}_{p}$ is a **convex set**.

(c) For fixed $a,b\in\mathbb{R}^{n}$ such that $a\neq b$, and fixed $k\in[0,1]$, the given set is
\begin{align*}
& \{x\in\mathbb{R}^{n} \mid \|x -a \|_2 \leq k \|x-b\|_2\}\\
=& \{x\in\mathbb{R}^{n} \mid \|x -a \|_2^2 \leq k^2 \|x-b\|_2^2\}\\
=& \{x\in\mathbb{R}^{n} \mid \left(1-k^{2}\right)x^{\top}x - 2\left(a-k^2 b\right)^{\top}x + \left(a^{\top}a-k^{2}b^{\top}b\right) \leq 0\},
\end{align*}
which for $k=0$, is a point ($x=a$). For $k=1$, the set becomes a halfspace. For $0<k<1$, the set becomes a 2-norm ball $\{x\in\mathbb{R}^{n}\mid (x-x_0)^{\top}(x-x_0)\leq r^2\}$ with center $x_0:=\frac{a-k^2 b}{1-k^2}$ and radius $r:= \sqrt{\frac{k^2\|b\|_2^2 - \|a\|_2^2}{1-k^2} - \|x_0\|_2^2}$. Since point, halfspace, ball--all are convex, the **given set is convex** for all $k\in[0,1]$.

(d) The given set $\{p\in\Delta^{n-1}\mid \alpha \leq \sum_{i=1}^{n}p_i f(a_i) \leq \beta\}$ is the intersection of two halfspaces with the standard simplex $\Delta^{n-1}$ (the standard simplex is a convex set from HW2, P1(c)(ii)), and hence being the intersection of three convex sets, is a **convex set**.

(e) By the definition of variance, the given set is $\{p\in\Delta^{n-1}\mid \sum_{i=1}^{n}p_i a_i^{2} - \left(\sum_{i=1}^{n}p_i a_i\right)^{2}\leq \alpha\}$ for fixed $\alpha>0$, which is the intersection of the standard simplex with a quadratic surface. As a simple counterexample, consider $n=2$, $a_1=0$, $a_2=1$, resulting in the set
$$\mathcal{X}:=\bigg\{\begin{pmatrix}
p_1\\
p_2
\end{pmatrix}\in\mathbb{R}^{2} \mid p_1,p_2\geq 0, \; p_1 + p_2 =1, \; p_2 - p_2^2 \leq \alpha\bigg\}.$$
Consider $p:=\begin{pmatrix}1\\
0\end{pmatrix}, q:=\begin{pmatrix}0\\
1\end{pmatrix}$. Clearly, both $p,q\in\mathcal{X}$ but their convex combination $\frac{1}{2}p + \frac{1}{2}q = \begin{pmatrix}1/2\\
1/2\end{pmatrix} \notin\mathcal{X}$ for $\alpha = 1/5$, for example. Therefore the given set is **nonconvex** in general.

# Problem 2 [15 + 15 + 15 + 5 = 50 points] A Set of Matrices

(a) Consider the set of matrices
$$\mathcal{X} := \{X\in\mathbb{R}^{n\times n} \mid X_{ij}\geq 0, \quad \sum_{i=1}^{n}X_{ij}=1, \quad \sum_{j=1}^{n}X_{ij}=1\quad\text{for all}\quad i,j=1,...,n\}.$$
**Prove** that $\mathcal{X}$ is a **convex set**.

(b) Is $\mathcal{X}$ a **cone**? **Why/why not?**

(c) **Prove** that $\mathcal{X}$ is a **polyhedron**. (Hint: review the definition of polyhedron from Lec. 6, p. 5)

(d) Fix $n=2$. In this case, **what is the relation** between $\mathcal{X}$ and the **convex hull of all $2\times 2$ permutation matrices** (which you encountered in HW2, Prob. 1(b))? **Explain briefly**.

# Solution for Prob. 2.

(a) Consider two matrices $X,Y\in\mathcal{X}$ and any $\theta\in[0,1]$. Then the matrix $Z := \theta X + (1-\theta)Y$ satisfies
$$Z_{ij}\geq 0, \quad \sum_{i=1}^{n}Z_{ij} = \theta\sum_{i=1}^{n}X_{ij} + (1-\theta)\sum_{i=1}^{n}Y_{ij} =\theta + (1-\theta)=1, \quad \sum_{j=1}^{n}Z_{ij}= \theta\sum_{j=1}^{n}X_{ij} + (1-\theta)\sum_{j=1}^{n}Y_{ij} =\theta + (1-\theta)=1\:\forall i,j=1,...,n,$$
and hence $Z= \theta X + (1-\theta)Y \in\mathcal{X}$ for any $\theta\in[0,1]$. Therefore, $\mathcal{X}$ is a **convex set**. 

Side remark: A matrix $X\in\mathcal{X}$ is callled doubly stochastic a.k.a. bistochastic matrix. Such matrices come up in the design of Markov chains.

(b) No, $\mathcal{X}$ is **not a cone** because for $X\in\mathcal{X}$, the nonnegative scaling $\theta X \notin\mathcal{X}$ for any $\theta\geq 0$. Specifically, the row and column sums of $\theta X$ are equal to $\theta$, as opposed to being unity. From the definition of $\mathcal{X}$, all row and column sums being exactly equal to unity are necessary for belonging to the set $\mathcal{X}$.

(c) From the definition of $\mathcal{X}$, this set is an intersection of $n^2$ halfspaces $X_{ij}\geq 0$, and $2n$ hyperplanes $\sum_{i=1}^{n}X_{ij}=1, \sum_{j=1}^{n}X_{ij}=1$ where $i,j=1,...,n$. Therefore, $\mathcal{X}$ **is a polyhedron** in $\mathbb{R}^{n^{2}}$ by definition (see Lec. 6, p. 5).

Side remark: $\mathcal{X}$ is often referred to as the [Birkhoff polytope](https://en.wikipedia.org/wiki/Birkhoff_polytope).

(d) From HW2, Prob. 1(b), the convex hull of all $2\times 2$ permutation matrices is 
$${\rm{conv}}\left(\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix},\begin{pmatrix}
0 & 1\\
1 & 0
\end{pmatrix}\right) = \bigg\{\theta\begin{pmatrix}
1 & 0\\
0 & 1
\end{pmatrix} + (1-\theta)\begin{pmatrix}
0 & 1\\
1 & 0
\end{pmatrix}\in\mathbb{R}^{2\times 2}, \quad 0\leq \theta\leq 1\bigg\} = \bigg\{\begin{pmatrix}
\theta & 1-\theta\\
1-\theta & \theta
\end{pmatrix}\in\mathbb{R}^{2\times 2}, \quad 0\leq \theta\leq 1\bigg\},$$
which is **exactly equal to $\mathcal{X} := \{X\in\mathbb{R}^{2\times 2} \mid X_{ij}\geq 0, \, \sum_{i=1}^{2}X_{ij}=1, \, \sum_{j=1}^{2}X_{ij}=1\,\text{for all}\, i,j=1,2\}$**, the set of $2\times 2$ elelement-wise nonnegative matrices with row and column sums equal to unity.

Side remark: This is true for general $n$ dimensions, not just for $n=2$. In words, the vertices of the Birkhoff polytope are the $n\times n$ permutation matrices, a result known as the Birkhoff-Von Neumann theorem.
