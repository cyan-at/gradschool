# Problem 1. [(1+9) x 5 = 50 points] Optimization Problems: Convex vs Nonconvex

For each of the following, **state** if the problem is nonconvex or convex. If your answer is nonconvex, then **explain why**. If your answer is convex, then **explain what type of convex optimization problem it is** among the following: LP, QP, QCQP, SOCP, SDP.

(a) $\underset{x\in\Delta^{n-1}}{\min} \langle c, x\rangle$ where $c\in\mathbb{R}^{n}$.

(b) $\underset{x\in\mathbb{R}^{n}}{\min} \langle c, x\rangle$ subject to $Ax \leq b$ where $A\in\mathbb{R}^{n\times n}$ nonsingular, and $b,c\in\mathbb{R}^{n}$.

(c) $\underset{M\in\mathbb{R}^{n\times n}}{\min}\;\langle C, M\rangle$ subject to $M\geq 0$ (elementwise), $M\boldsymbol{1}=a$, $M^{\top}\boldsymbol{1}=b$ where $a,b\in\mathbb{R}^{n}_{+}$, $C\in\mathbb{R}^{n\times n}$, and $\boldsymbol{1}\in\mathbb{R}^{n}$ denotes a vector of ones.

(d) $\underset{x\in\mathbb{R}^{n}}{\min}\;x^{\top}P x$ subject to $x^{\top}Q x \leq 1$ where $P\in\mathbb{S}^{n}$, $Q\in\mathbb{S}^{n}_{++}$.

(e) $\underset{(x,y,z)\in\mathbb{R}^{n}\times\mathbb{R}_{+}\times\mathbb{R}_{+}}{\min} \bigg\{\langle c, x\rangle + \alpha y + \beta z\bigg\}$ subject to $x^{\top}x\leq yz$ where $c\in\mathbb{R}^{n}$, $\alpha,\beta\in\mathbb{R}$.

## Solution for Prob. 1

(a) This is a **convex optimization problem**. By Lec. 11, p. 4, this is an **LP** because the objective is linear in decision variable $x$, and the constraint set being standard simplex, is a polyhedron.

(b) This is a **convex optimization problem**. By Lec. 11, p. 4, this is an **LP** because the objective is linear in decision variable $x$, and the constraint set being an intersection of $n$ closed halfspaces, is a polyhedron.

(c) This is a **convex optimization problem**. By Lec. 11, p. 4, this is an **LP** because the objective is linear in decision variable $M$ (i.e., linear in entries of $M$), and the constraint set being an intersection of $n^2$ closed halfspaces and $2n$ hyperplanes, is a polyhedron.

(d) This is a **nonconvex optimization problem** because the objective is nonconvex in general for $P\in\mathbb{S}^{n}$.

(e) This is a **convex optimization problem**. To understand what kind of convex optimization problem it is, we first note that the objective function is linear in the decision variable $(x,y,z)\in\mathbb{R}^{n}\times\mathbb{R}_{+}\times\mathbb{R}_{+}$. The constraints $y\geq 0, z \geq 0$ define intersection of two closed halfspaces, i.e., a polyhedron. The feasible set is an intersection of this polyhedron with $\{(x,y,z)\mid x^{\top}x \leq yz\}$. Notice that
\begin{align*}
x^{\top}x \leq yz \;\Leftrightarrow\; 4x^{\top}x \leq 4yz = (y+z)^{2} - (y-z)^{2} \;\Leftrightarrow\; 4x^{\top}x + (y-z)^{2} \leq (y+z)^{2} \;\Leftrightarrow\; \bigg\Vert\begin{pmatrix}
2x\\
y-z
\end{pmatrix}\bigg\Vert_{2} \leq y+z,
\end{align*}
which is a composition of second order cone constraint with affine map. Therfore, by Lec. 11, p. 8, this is an **SOCP**.

## Solution for Prob. 2

(i) Let us sort the entries of $c$ as $c_1 = c_2 = ... = c_{k} < c_{k+1} \leq ... \leq c_{n}$. Then, $\langle c, x\rangle \geq \langle c_{1}\boldsymbol{1},x\rangle = c_{1}\left(\boldsymbol{1}^{\top}x\right) = c_{1}$ for any $x\in\Delta^{n-1}$, with equality achieved by all vectors $x^{\rm{opt}} \in\Delta^{n-1}$ such that $x_{1}^{\rm{opt}} + ... + x_{k}^{\rm{opt}} = 1$, and $x_{k+1}^{\rm{opt}} = ... = x_{n}^{\rm{opt}} = 0$. The minimum value is $c_1$.

(ii) Introducing the change-of-variable $y:=A^{-1}x$, the problem becomes: $\underset{y\in\mathbb{R}^{n}}{\min}c^{\top}A^{-1}y$ subject to $y \preceq b$. Therefore,
$$\underset{y\in\mathbb{R}^{n}}{\min}c^{\top}A^{-1}y = \begin{cases}
c^{\top}A^{-1}b & \text{if}\quad A^{-\top}c \preceq 0\;\text{(then the minimizer is $y^{\rm{opt}}=b$)},\\
-\infty & \text{otherwise}.
\end{cases}$$

(iii) Because $Q\in\mathbb{S}^{n}_{++}$, it has unique square root $Q^{1/2}\in\mathbb{S}^{n}_{++}$. Introducing the change-of-variable $y:=Q^{1/2}x$, the problem becomes:
$$\underset{y^{\top}y \leq 1}{\min}y^{\top}Q^{-1/2}PQ^{-1/2}y.$$
Because $y=0$ is in the feasible set, the above minimum value must be $\leq 0$.

Suppose if possible that there exists a nonzero gloabl minimizer $y^{\rm{opt}} \neq 0$ such that $0< \|y^{\rm{opt}}\|_{2} =: \delta < 1$, i.e., the nonzero gloabl minimizer $y^{\rm{opt}}$ lies in the interior of the feasible set. Now let $\tilde{y} := y^{\rm{opt}}/\delta$. Since $\tilde{y}^{\top}\tilde{y} = \frac{1}{\delta^2}\|y^{\rm{opt}}\|_{2}^{2} = 1$, the vector $\tilde{y}$ is feasible. However, the objective evaluated at $\tilde{y}$ becomes
$$\tilde{y}^{\top}Q^{-1/2}PQ^{-1/2}\tilde{y} = \underbrace{\frac{1}{\delta^2}}_{>1}\underbrace{\left(y^{\rm{opt}}\right)^{\top}Q^{-1/2}PQ^{-1/2}y^{\rm{opt}}}_{\leq 0} \leq \left(y^{\rm{opt}}\right)^{\top}Q^{-1/2}PQ^{-1/2}y^{\rm{opt}},$$
which contradicts the supposition that the nonzero global minimizer is in the interior of the unit ball. Therefore, if a nonzero global minimizer exists, then it must be on the boundary of the feasible set (i.e., must be a unit vector). 

But linear algebra (also in Lec. 9, p. 12-13) tells that if $y^{\top}y=1$, then the minimum value is $\lambda_{\min}\left(Q^{-1/2}PQ^{-1/2}\right)$. We conclude that
\begin{align*}
&\underset{\{x\in\mathbb{R}^{n}\mid \,x^{\top}Qx\leq 1,\, Q\in\mathbb{S}^{n}_{++}\}}{\min}\;x^{\top}P x\\
&=\underset{y^{\top}y \leq 1}{\min}y^{\top}Q^{-1/2}PQ^{-1/2}y \\
&= \begin{cases}
\lambda_{\min}\left(Q^{-1/2}PQ^{-1/2}\right) & \text{if}\quad \lambda_{\min}\left(Q^{-1/2}PQ^{-1/2}\right) \leq 0\;\text{(then the minimizer is normalized eigenvector corresponding to this $\lambda_{\min}$)},\\
0 & \text{otherwise}\;\text{(then the minimizer is $y=0$)}.
\end{cases}
\end{align*}
