# Problem 2 [15 + 15 + 15 + 5 = 50 points] A Set of Matrices

a
1. This follows from the $\theta$ definition of convexity. Given $A, B \in \mathcal{X}, \theta \in [0, 1]$, define $C$ and show it to be also belong to $\mathcal{X}$:
\begin{align}
C & = \theta A + (1-\theta)B \\
C_{ij} &= \theta A_{ij} + (1-\theta) B_{ij} = \theta (\geq 0) + (1-\theta) (\geq 0) \geq 0 \\
\Sigma_i^{n} C_{ij} &= \Sigma_i^{n} \theta A_{ij} + \Sigma_i^{n} (1-\theta) B_{ij} = \theta \Sigma_i^{n} A_{ij} + (1-\theta) \Sigma_i^{n} B_{ij} = \theta + (1-\theta) = 1 \\
\Sigma_j^{n} C_{ij} &= \Sigma_j^{n} \theta A_{ij} + \Sigma_j^{n} (1-\theta) B_{ij} = \theta \Sigma_j^{n} A_{ij} + (1-\theta) \Sigma_j^{n} B_{ij} = \theta + (1-\theta) = 1 \\
\end{align}
2. Then from above, we show that for **any** $\theta \in [0, 1]$ and **any** $A, B \in \mathcal{X}$, the interpolate matrix $C = \theta A + (1-\theta)B$ satisfies all 3 properties of a matrix in $\mathcal{X}$, belongs also in $\mathcal{X}$
3. Therefore, the set $\mathcal{X}$ **is convex**

b
1. $\mathcal{X}$ **is not** a cone, because it for a non-negative scalar $\theta \geq 0$, i.e. $\theta = 0.5$, $C = \theta A$ for $A \in \mathcal{X}$:
\begin{align}
\Sigma_i^{n} C_{ij} = \Sigma_i^{n} \theta A_{ij} = \theta \Sigma_i^{n} A_{ij} = \theta = 0.5 \neq 1
\end{align}
2. Then see that for this counterexample, the non-negative scaling of $A$ results in a matrix $ \notin \mathcal{X}$, then $\mathcal{X}$ is **not a cone**

c
1. You can rearrange the matrix $X \in \mathcal{X}$ as a vector in $\mathbb{R}^{n^2}$ by concatenating every column together. Let's call this vector $\underline{X}$
2. We can define a matrix $C \in \mathbb{R}^{nxn^2}$ such that:
\begin{equation}
C_{ij} = 
\left\{
    \begin{array}{lr}
        1, & \text{for } i=1...n, j=(i-1)n + k \quad \text{for } k=1...n \\
        0, & \text{otherwise } 
    \end{array}
\right\}
\end{equation}
3. For example, for $n=2$, such a vector $\underline{X}$ and $C$ matrix looks like so:
\begin{align}
X_{n=2} & = \begin{bmatrix} a & b \\ c & d \end{bmatrix} \\
\underline{X_{n=2}} & = \begin{bmatrix} a \\ c \\ b \\ d \end{bmatrix} \\
C_{n=2} & = \begin{bmatrix} 1 & 1 & 0 & 0 \\ 0 & 0 & 1 & 1 \end{bmatrix} \\
\end{align}
4. Then from above we see that the constraint that $\Sigma_{i=1}^{n} X_{ij} = 1$ can be expressed as the equality / hyperplane:
\begin{align}
C_{n=2} \underline{X_{n=2}} = \begin{bmatrix} a + c \\ b + d \end{bmatrix} = 1
\end{align}
5. And similar to $C$ we can define a selection matrix $R$ to do the same:
\begin{equation}
R_{ij} = 
\left\{
    \begin{array}{lr}
        1, & \text{for } i=1...n, j=i+nk \quad \text{for } k=0...n-1 \\
        0, & \text{otherwise } 
    \end{array}
\right\}
\end{equation}
\begin{align}
R_{n=2} & = \begin{bmatrix} 1 & 0 & 1 & 0 \\ 0 & 1 & 0 & 1 \end{bmatrix} \\
R_{n=2} \underline{X_{n=2}} & = \begin{bmatrix} a + b \\ c + d \end{bmatrix} = 1
\end{align}
6. So we see that the row and column wise sum constraints in $\mathcal{X}$ can be expressed as $2n$ equalities, each involving $n^2$ terms.
7. We can also see that the inequality constraints $0 \leq X_{ij} \leq 1$ can be expressed as $X_{ij} \leq 1$ and $-X_{ij} \leq 0$. And we can create another such selection matrix $P \in \mathbb{R}^{2n^2xn^2}$ where the top half is the identity matrix $I_{n^2}$ and bottom half is $-I_{n^2}$, and the b vector $\underline{b}$ on the RHS of the inequality is top half 1, bottom half 0. 
8. Then the bounds constraint can be expressed as $2n^2$ halfspaces / inequality constraints of the form $P \underline{X} \leq \underline{b}$
9. From 1-8 above, we see we can rewrite the requirements defining the set $\mathcal{X}$ as a set of $2n$ hyperplanes and $2n^2$ halfspaces. Then by definition from the lecture note, such a set is a **polyhedron**

\end{align}
2. The convex **hull** $P_{n=2}$ is by definition the set:
\begin{align}
\mathcal{C} = \text{conv}(P_{n=2}) = \{ \theta A + (1-\theta) B \quad \text{for } \theta \in [0, 1] \}
\end{align}
3. Then we see that for any column sum of $\mathcal{C}$, $C$:
\begin{align}
\Sigma_{i=1}^{n} C_{ij} & = \theta \Sigma_{i=1}^{n} A_{ij} + (1-\theta) \Sigma_{i=1}^{n} B_{ij} \\
& = \theta + (1-\theta)
& = 1
\end{align}
4. The same reasoning above applies to every **row sum** of $C \in \mathcal{C}$
5. Finally, note that since every element of $A, B$ are non-negative, and $\theta, 1-\theta$ are both non-negative, then every element in any $C \in \mathcal{C}$ is also non-negative
6. Then every element of $\mathcal{C}$ satisfies every criteria of, is also an element of $\mathcal{X}$
