# Problem 1 [50 points] Computer-controlled continuous-time LTI system

Consider a continuous-time linear time invariant (LTI) system 

$$\dot{x} = A x + B u, \quad x\in\mathbb{R}^{n},\quad u\in\mathbb{R}^{m}, \quad x(t=0)=x_{0},$$

where a digital computer supplies the control signal $u(t)$ with sampling period $h>0$. Consequently, 

$$u(t) \equiv u(k) := u(t=kh)\quad\forall t\in[kh,(k+1)h), \quad k=0,1,2,...$$

This makes the control input $u$ a piecewise constant function of the time $t$.

## (a) [20 points] Equivalent discrete-time LTI system

**Prove that** the control system above is equivalent to a discrete-time LTI system of the form

$$x(k+1) = A_{\rm{d}} x(k) + B_{\rm{d}} u(k), \quad x(k) = x_{0}, \quad k=0,1,2,...$$

In other words, derive the matrices $A_{\rm{d}}\in\mathbb{R}^{n\times n}, B_{\rm{d}}\in\mathbb{R}^{n\times m}$ as function of the original system matrices $A,B$ and the sampling period $h>0$.

(**Hint:** The solution of the ODE $\dot{x}=Ax+Bu$ with initial condition $x(t=0)=x_{0}$, is $x(t) = e^{At}x_{0} + \int_{0}^{t}e^{A(t-\tau)}B u(\tau){\rm{d}}\tau$. Then use the fact that $x(k):=x(t=kh)$ etc.)

1. Let $x[k] = x(kh)$, $x[k+1] = x((k+1)h)$
  
2. $x[k] = e^{Akh}x(0) + \int_{0}^{kh}e^{A(kh-\tau)}Bu(\tau){\rm{d}}\tau$
  
3.  
$
\begin{align}
&x[k+1] \\
&= e^{A(k+1)h}x(0) + \int_{0}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&= e^{A(k+1)h}x(0) + \int_{0}^{kh}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau + \int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&= e^{A(k+1)h}x(0) + \int_{0}^{kh}e^{Ah}e^{A(kh-\tau)}Bu(\tau){\rm{d}}\tau + \int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&= e^{A(k+1)h}x(0) + e^{Ah}\int_{0}^{kh}e^{A(kh-\tau)}Bu(\tau){\rm{d}}\tau + \int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&= e^{Ah}e^{Akh}x(0) + e^{Ah}\int_{0}^{kh}e^{A(kh-\tau)}Bu(\tau){\rm{d}}\tau + \int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&= e^{Ah} \left[ e^{Akh}x(0) + \int_{0}^{kh}e^{A(kh-\tau)}Bu(\tau){\rm{d}}\tau \right] + \int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&= e^{Ah}x[k] + \int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
\end{align}
$
  
4. Let $v(\tau) = (k+1)h - \tau$, $\rm{d}v = -\rm{d}\tau \Rightarrow \rm{d}\tau = -\rm{d}v$
  
5. Assume $u(\tau)$ in $kh \Leftrightarrow (k+1)h$ is constant $\Rightarrow \int_{kh}^{(k+1)h}Bu(\tau){\rm{d}}\tau = Bu[k]$
  
6. then: 
$
\begin{align}
&\int_{kh}^{(k+1)h}e^{A((k+1)h-\tau)}Bu(\tau){\rm{d}}\tau \\
&=-Bu[k]\int_{v(kh)}^{v((k+1)h)}e^{Av}{\rm{d}}v \\
&=-Bu[k]\int_{h}^{0}e^{Av}{\rm{d}}v \\
&=Bu[k]\int_{0}^{h}e^{Av}{\rm{d}}v \\
=A^{-1}(e^{Ah}-I)Bu[k]
\end{align}
$
  
7. Combining 3. and 6. we get (if A nonsingular, $A^{-1}$ exists:

\begin{gather*} 
x[k+1] = e^{Ah}x[k] + A^{-1}(e^{Ah} - I)B u[k] \\
A_{d} = e^{Ah} \\
B_{d} = A^{-1}(e^{Ah}-I)B
\end{gather*}
  
8. If A is singular, replace w/ Taylor expansion:

\begin{gather*} 
e^{Ah} = \sum_{k=0}^\infty \frac{1}{k!}(Ah)^{k} = A_{d} \\
B\int_{0}^{h}e^{Av}{\rm{d}}v = B(h(I + \frac{Ah}{2!} + ... \frac{(Ah)^{k-1}}{k!}) = B\sum_{k=1}^\infty \frac{1}{k!}A^{k-1}h^{k} = B_{d}
\end{gather*}

# Problem 2 [50 points] Lyapunov analysis in discrete-time

For discrete-time autonomous nonlinear system $x(k+1)=f(x(k))$, $k=0,1,2,...$, one can derive a Lyapunov stability theorem analogous to the continuous-time case, by simply replacing the condition $\dot{V}<$ (or $\leq$) $0$ to its discrete-time counterpart: $V(k+1)<$ (or $\leq$) $V(k)$, where $V(k):=V(x(k))$, while keeping the other conditions (positive definiteness/semi-definiteness) same. With this modification, related theorems such as Barbashin-Krasovskii and LaSalle apply.

Consider the following 2D nonlinear system

$$x_{1}(k+1) = \dfrac{\alpha x_{2}(k)}{1 + \left(x_{1}(k)\right)^{2}}, \qquad x_{2}(k+1) = \dfrac{\beta x_{1}(k)}{1 + \left(x_{2}(k)\right)^{2}}, \qquad 0< \alpha^2 < 1, \quad 0 < \beta^2 < 1.$$
  


## (a) [5 + 15 = 20 points] Fixed point

(i) **Prove that** origin is a fixed point. 
  
1. Let $x_{1}[k] = 0, x_{2}[k] = 0$
  
2. Then $x_{1}[k+1] = \frac{\alpha (0)}{1 + 0^{2}} = 0 = x_{1}[k] $
  
3. Then $x_{2}[k+1] = \frac{\beta (0)}{1 + 0^{2}} = 0 = x_{2}[k] $
  
4.  So at $[0, 0]$, $x_{1}[k+1] = x_{1}[k], x_{2}[k+1] = x_{2}[k]$, which by definition is a fixed point


(ii) **Prove that** origin is in fact the **unique** fixed point.
  
1. Let $u = x_{1}[k], v = x_{2}[k]$

2. For a fixed point, $x_{1}[k+1] = x_{1}[k] = u$, so:
  
$
\begin{align}
&u = \frac{\alpha v}{1 + (u)^{2}} \\
&v = \frac{1}{\alpha}{(u + u^{3})} \\
&\frac{1}{\alpha}{(u + u^{3})} = \frac{\beta u}{1 + (\frac{1}{\alpha}{(u + u^{3})})^{2}} \\
&\frac{1}{\alpha}{(1 + u^{2})} = \frac{\beta \alpha^{2}}{\alpha^{2} + (u + u^{3})^{2}}  \\
&\frac{1}{\alpha} + \frac{u^{2}}{\alpha} + \frac{u^{2}}{\alpha^{3}} + \frac{3u^{4}}{\alpha^{3}} + \frac{3u^{6}}{\alpha^{3}} + \frac{u^{8}}{\alpha^{3}} = \beta
\end{align}
$
  
3. Let $k = \frac{u^{2}}{\alpha} + \frac{u^{2}}{\alpha^{3}} + \frac{3u^{4}}{\alpha^{3}} + \frac{3u^{6}}{\alpha^{3}} + \frac{u^{8}}{\alpha^{3}}$
  
4.
$
\begin{align}
&u \neq 0 \Rightarrow k > 0 \\
&\Rightarrow \frac{1}{\alpha} + k > \beta \\
&\Rightarrow k = \beta - \frac{1}{\alpha} > 0 \\
&\Rightarrow \beta > \frac{1}{\alpha} \\
&\Rightarrow \alpha \beta > 1 \\
&\Rightarrow \alpha^{2} \beta^{2} > 1 \\
\end{align}
$
  
5. But we know $\alpha^{2} < 1, \beta^{2} < 1 \Rightarrow \alpha^{2}\beta^{2} < 1$, so 4. is a contradiction, therefore, $u = 0$
  
6.
$
\begin{align}
&u = 0 \Rightarrow x_{1}[k] = 0 \\
&\Rightarrow x_{2}[k] = \frac{1}{\alpha}(0 + 0) = 0
\end{align}
$

so we show that if $u, v$ is a fixed point, $u, v = [0, 0]$

## (b) [20 points] AS

**Prove that** the origin is asymptotically stable (AS).
  
1. Let $V(k) = \frac{1}{2} \left[ x_{1}[k]^{2} + x_{2}[k]^{2} \right] = \frac{1}{2} \begin{bmatrix}[ x_{1}[k] & x_{2}[k] \end{bmatrix} \left( P \right) \begin{bmatrix} 1 & 0\\ 0 & 1 \end{bmatrix} \begin{bmatrix}[ x_{1}[k] \\ x_{2}[k] \end{bmatrix} $
  
2. We know $P$ is symmetric positive definite, so $V(k)$ is a positive definite function.
  
3.
$
\begin{align}
&V(k+1) = \frac{1}{2} \left[ x_{1}[k+1]^{2} + x_{2}[k+1]^{2} \right] \\
&=\frac{1}{2} \left[ \left( \frac{\alpha x_{2}[k]}{1 + x_{1}[k]^{2}}  \right)^{2} + \left( \frac{\beta x_{1}[k]}{1 + x_{2}[k]^{2}}  \right)^{2} \right] \\
&=\frac{1}{2} \left[ \left( \alpha^{2} \frac{1}{(1 + x_{1}[k]^{2})^{2}} x_{2}[k]^{2}  \right) + \left( \beta^{2} \frac{1}{(1 + x_{2}[k]^{2})^{2}} x_{1}[k]^{2}  \right) \right] \\ 
&=\frac{1}{2} \left[ \alpha^{2} C x_{2}[k]^{2} + \beta^{2} D x_{1}[k]^{2} \right], C = \frac{1}{(1 + x_{1}[k]^{2})^{2}}, D = \frac{1}{(1 + x_{2}[k]^{2})^{2}} \\
\end{align} 
$
  
4. We can see that $0 < C <= 1$, $0 < D <= 1$ since both are reciprocals of the sum of $1 + x$ where $x >= 0$
  
5.
$
\begin{align}
&0 < \alpha^{2} < 1 \\
&\Rightarrow 0 < \alpha^{2} C < 1 \\
&\Rightarrow 0 < \alpha^{2} C \left( \frac{1}{2} x_{2}[k]^{2} \right) < \frac{1}{2} x_{2}[k]^{2} \\
\end{align}
$
  
6.
$
\begin{align}
&0 < \beta^{2} < 1 \\
&\Rightarrow 0 < \beta^{2} D < 1 \\
&\Rightarrow 0 < \beta^{2} D \left( \frac{1}{2} x_{1}[k]^{2} \right) < \frac{1}{2} x_{1}[k]^{2} \\
\end{align}
$
  
7.
$
\begin{align}
&V(k+1) = \alpha^{2} C \left( \frac{1}{2} x_{2}[k]^{2} \right) + \beta^{2} D \left( \frac{1}{2} x_{1}[k]^{2} \right) < \frac{1}{2} x_{1}[k]^{2} + \frac{1}{2} x_{2}[k]^{2} = V(k)\\
&\Rightarrow V(k+1) < V(k)\\
\end{align}
$  
Therefore, $V(k)$ is a negative definite function.

8. Combining $2.$ and $7.$, from Lecture 3 pg 3, $x = 0$ is AS.
