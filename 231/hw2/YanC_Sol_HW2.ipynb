# Problem 1 [40 points] Lyapunov stability for a non-autonomous nonlinear system

Consider the continuous-time non-autonomous system

$$\begin{align*}
\dot{x}_{1} &= -x_{1} - g(t)x_{2},\\
\dot{x}_{2} &= x_{1} - x_{2},
\end{align*}
$$

where the state vector $(x_1,x_2)^{\top}\in\mathbb{R}^{2}$. The function $g(t)$ is $C^{1}$ in time $t$, satisfies $0 \leq g(t) \leq k$ for some constant $k$, and $\dot{g}(t) \leq g(t)$ for all $t\geq 0$.

## (a) [10 points] Fixed point

**Prove that** the origin $(0,0)^{\top}$ is a fixed point. 

1.
$$\begin{align*}
&\forall t \geq 0,\\
\dot{x}_{1} &= -0 - g(t) * 0 &= 0,\\
\dot{x}_{2} &= 0 - 0 &= 0,
\end{align*}
$$

Thus by definition $(0,0)^{\top}$ is a fixed point

## (b) [30 points] Origin is GES

Use the Lyapunov function $V(t,x_1,x_2) := x_1^2 + (1 + g(t))x_2^2$ to **prove that** the origin is globally exponentially stable (GES).

(**Hint:** Lec. 5, p. 3).

1.
\begin{align*}
0 \leq x_{2}^{2} \leq x_{2}^{2} + x_{1}^{2}\\
0 \leq x_{2}^{2} \leq \lVert x \rVert^{2}\\
0 \leq g(t) x_{2}^{2} \leq k \lVert x \rVert^{2}\\
\lVert x \rVert^{2} \leq g(t) x_{2}^{2} + \lVert x \rVert^{2} \leq (1+k) \lVert x \rVert^{2}\\
\lVert x \rVert^{2} \leq g(t) x_{2}^{2} + x_{2}^{2} + x_{1}^{2} \leq (1+k) \lVert x \rVert^{2}\\
\lVert x \rVert^{2} \leq (g(t) + 1) x_{2}^{2} + x_{1}^{2} \leq (1+k) \lVert x \rVert^{2}\\
\lVert x \rVert^{2} \leq V(t, x_{1}, x_{2}) \leq (1+k) \lVert x \rVert^{2}\\
\end{align*}
<br>
2. We take 1. above $k_{1} = 1, k_{2} = (1 + k), \alpha = 2$:
\begin{align*}
k_{1} \lVert x \rVert^{\alpha} \leq V(t, x_{1}, x_{2}) \leq k_{2} \lVert x \rVert^{\alpha}
\end{align*}
<br>
3. 
\begin{align}
\frac{\partial{V}}{\partial{t}} & = 0 + 0 + \frac{\partial{}}{\partial{t}} g(t)x_{2}^{2}\\
& = x_{2}^{2}\dot{g}(t)
\end{align}
<br>
4.
\begin{align}
\frac{\partial{V}}{\partial{x}} & = \begin{bmatrix} \frac{\partial{V}}{\partial{x_{1}}} \\ \frac{\partial{V}}{\partial{x_{2}}} \end{bmatrix}\\
& = \begin{bmatrix} 2x_{1} \\ 2x_{2}(1+g) \end{bmatrix}\\
\end{align}
<br>
5. Combining 3. 4. above:
\begin{align}
\dot{V} & = \frac{\partial{V}}{\partial{t}} + \frac{\partial{V}}{\partial{x}} f(t, x_{1}, x_{2})\\
& = x_{2}^{2}\dot{g}(t) + 2x_{1}(-x_1 - gx_2) + (x_1 - x_2)(2x_2)(1+g)\\
& = -2x_1^{2} + 2x_1x_2 + x_2^{2}(\dot{g} -2 - 2g)\\
\end{align}
<br>
6. Since  $\dot{g} \leq g$:
\begin{align}
\dot{g} - 2 - 2g \leq g - 2 - 2g = -2 - g\\
\end{align}
and further $-2 - g \leq -2$:
\begin{align}
0 \leq g \leq k\\
-k \leq -g \leq 0\\
-2-k \leq -2-g \leq -2\\
\end{align}
therefore $\dot{g} - 2 - 2g \leq -2$
<br>
7. Combinining 5. and 6. above: $\dot{V} \leq  2x_1x_2 - 2x_1^{2} - 2x_2^{2}$
<br>
8. $(x_1 - x_2)^{2}$ is non-negative, so:
\begin{align}
\dot{V} & \leq  2x_1x_2 - 2x_1^{2} - 2x_2^{2}\\
& \leq 2x_1x_2 - 2x_1^{2} - 2x_2^{2} + (x_1 - x_2)^{2}\\
& = 2x_1x_2 - 2x_1^{2} - 2x_2^{2} + x_1^{2} - 2x_1x_2 + x_2^{2}\\
& = - 2x_1^{2} - 2x_2^{2} + x_1^{2} + x_2^{2}\\
& = - x_1^{2} - x_2^{2}\\
& = -1 (x_1^{2} + x_2^{2})\\
& = -1 \lVert x \rVert^{2}\\
& \leq k_{3} \lVert x \rVert^{\alpha}, k_{3} = 1, \alpha = 2\\
\end{align}
<br>
9. So from  8. and 2. since they hold true $\forall x \in {R^{n}}$ and by Lec 5 pg. 3, the origin **is GES**.

# Problem 2 [60 points] Lyapunov stability for a rotating rigid spacecraft

The controlled dynamics for a rotating rigid spacecraft is given by the Euler equation

\begin{align*}
J_{1}\dot{\omega}_{1} = \left(J_{2} - J_{3}\right)\omega_{2}\omega_{3} + \tau_{1},\\
    J_{2}\dot{\omega}_{2} = \left(J_{3} - J_{1}\right)\omega_{3}\omega_{1} + \tau_{2},\\
    J_{3}\dot{\omega}_{3} = \left(J_{1} - J_{2}\right)\omega_{1}\omega_{2} + \tau_{3},
\end{align*}
where the parameters $J_1,J_2,J_3>0$ denote the principal moments of inertia; the state vector $(\omega_1,\omega_2,\omega_3)^{\top}\in\mathbb{R}^{3}$ denotes the spacecraftâ€™s angular velocity (in rad/s) along its principal axes; and the control vector $(\tau_1,\tau_2,\tau_3)^{\top}$ denotes the torque input applied about the principal axes.

## (a) [2 + (2 + 6) = 10 points] Fixed points in the absence of control

Suppose that the controls $\tau_1 = \tau_2 = \tau_3 \equiv 0$.

(i) **Argue that** origin is a fixed point.

1. when $(\omega_{1}, \omega_{2}, \omega_{3})^{T} = 0$

\begin{align*}
J_{1}\dot{\omega}_{1} = \left(J_{2} - J_{3}\right)0+ 0 \rightarrow J_{1}\dot{\omega}_{1} = 0 \rightarrow \dot{\omega}_{1} = 0,\\
J_{2}\dot{\omega}_{2} = \left(J_{3} - J_{1}\right)0 + 0 \rightarrow J_{2}\dot{\omega}_{2} = 0 \rightarrow \dot{\omega}_{2} = 0,\\
J_{3}\dot{\omega}_{3} = \left(J_{1} - J_{2}\right)0 + 0 \rightarrow J_{3}\dot{\omega}_{3} = 0 \rightarrow \dot{\omega}_{3} = 0,
\end{align*}

(ii) **How many fixed points other than origin** are there? **Explain what physical motions** do these non-origin fixed points correspnd to?

1. All 3 equations above include 2 / 3 state terms. So for any states where 2 of the 3 dimensions of $\omega_{i} = 0$, all 3 equations will yield 0. i.e. $(\omega_{1}, 0, 0)^{T}, (0, \omega_{2}, 0)^{T}, (0, 0, \omega_{3})^{T}$

2. For the nonzero elements of those 3 state vectors, it doest *not* matter what the nonzero element is, so there is an *infinite* # of points other than the origin.

3. The physical meaning of those 3 state vectors is when the spacecraft is rotating about *1* of the principal axes while stationary about the other axes. In these states, the spacecraft dynamics do *not* change, the spacecraft will continue rotating about that axis at the same angular velocity indefinitely.

## (b) [(10 + 5) + 10 = 25] S/AS in the absence of control

As in part (a), assume that the controls $\tau_1 = \tau_2 = \tau_3 \equiv 0$.

(i) By constructing a suitable Lyapunov function, **prove that** the origin is stable (S). From this analysis, **what can you conclude** about the asymptotic stability (AS) of the origin?

1. Let $V = \frac{1}{2}\begin{bmatrix} \omega_{1} & \omega_{2} & \omega_{3} \end{bmatrix} \begin{bmatrix} J_{1} & 0 & 0 \\ 0 & J_{2} & 0 \\ 0 & 0 & J_{3} \end{bmatrix} \begin{bmatrix} \omega_{1} \\ \omega_{2} \\ \omega_{3} \end{bmatrix} = \frac{1}{2}(J_{1}\omega_{1}^{2} + J_{2}\omega_{2}^{2} + J_{3}\omega_{3}^{2})$ where $P = \begin{bmatrix} J_{1} & 0 & 0 \\ 0 & J_{2} & 0 \\ 0 & 0 & J_{3} \end{bmatrix}$
<br>
2. Knowing $J_{1}, J_{2}, J_{3} > 0$, $V = \frac{1}{2}(J_{1}\omega_{1}^{2} + J_{2}\omega_{2}^{2} + J_{3}\omega_{3}^{2}) > 0$ $\forall \begin{bmatrix} \omega_{1} & \omega_{2} & \omega_{3} \end{bmatrix} \rightarrow$ P is positive definite matrix.
<br>
3. We can show that $\dot{V}$ is negative **semi-definite**:
\begin{align}
& \dot{V} = \omega_{1}(J_{1}*\dot\omega_{1}) + \omega_{2}(J_{2}*\dot\omega_{2}) + \omega_{3}(J_{3}*\dot\omega_{3}) \\
& = \omega_{1}(J_{2}-J_{3})\omega_{2}\omega_{3} + \omega_{2}(J_{3}-J_{1})\omega_{1}\omega_{3} + \omega_{3}(J_{1}-J_{2})\omega_{1}\omega_{1} \\
& = \omega_{1}\omega_{2}\omega_{3}(J_{2}-J_{3} + J_{3}-J_{1} + J_{1}-J_{2})\\
& = \omega_{1}\omega_{2}\omega_{3}(J_{2}-J_{3} + J_{3}-J_{1} + J_{1}-J_{2})\\
& = 0\\
\end{align}
<br>
4. Therefore, from 2. and 3. above, and by Lec 3. pg 2 pg 3, we prove that the origin **is S, not AS**

(ii) Given initial conditions $(\omega_{10},\omega_{20},\omega_{30})^{\top}$ and parameters $J_1 \neq J_2 \neq J_3$, it is possible to explicitly solve for the flow (still assuming $\tau_1 = \tau_2 = \tau_3 = 0$) as
\begin{align*}

The functions $\mathrm{cn},\mathrm{sn},\mathrm{dn}$ are the so-called Jacobi elliptic functions and are periodic in time $t$. **Using this information alone, and ignoring the previous Lyapunov analysis, explain what can you conclude** about the asymptotic stability of the origin?

1. If the $\omega_{i}(t)$ terms are **periodic** then by definition they by definition they do **not** converge to any one value i.e. at some starting time and period $t_{1}, p_{1}$, $\omega_{i}(t_{1}) = \omega_{i}(t_{1} + k_{1}p_{1}) = \alpha$ for $k_{1} = 1, 2, 3, ... \infty$ and for some other $t_{2}, p_{2}$, $\omega_{i}(t_{2}) = \omega_{i}(t_{2} + k_{2}p_{2}) = \beta$ for $k_{2} = 1, 2, 3, ... \infty$ s.t. $\alpha \neq \beta$
<br>
2. So if $\omega_{i}(t)$ converges, then that means $\lim_{t \to +\infty} w_{i}(t) = w_{i}(t_{1} + k_{1}p_{1}) = w_{i}(t_{2} + k_{2}p_{2}) = \gamma$ but this contradicts the statement in 1. So $\omega_{i}(t)$ **does not converge**

3. Thus by definition, given $\omega_{i}(t)$ the origin is **not AS**

## (c) [25 points] Global asymptotic stabilization using feedback control

For $i=1,2,3$, consider the state feedback control law $\tau_i = -k_i\omega_i$, where $k_i>0$ are constants. **Prove that** origin of the closed-loop system is globally asymptotically stable (GAS).

1. Let $V = \frac{1}{2}\begin{bmatrix} \omega_{1} & \omega_{2} & \omega_{3} \end{bmatrix} \begin{bmatrix} J_{1} & 0 & 0 \\ 0 & J_{2} & 0 \\ 0 & 0 & J_{3} \end{bmatrix} \begin{bmatrix} \omega_{1} \\ \omega_{2} \\ \omega_{3} \end{bmatrix} = \frac{1}{2}(J_{1}\omega_{1}^{2} + J_{2}\omega_{2}^{2} + J_{3}\omega_{3}^{2})$ where $P = \begin{bmatrix} J_{1} & 0 & 0 \\ 0 & J_{2} & 0 \\ 0 & 0 & J_{3} \end{bmatrix}$
<br>
2. Knowing $J_{1}, J_{2}, J_{3} > 0$, $V = \frac{1}{2}(J_{1}\omega_{1}^{2} + J_{2}\omega_{2}^{2} + J_{3}\omega_{3}^{2}) > 0$ $\forall \begin{bmatrix} \omega_{1} & \omega_{2} & \omega_{3} \end{bmatrix} \rightarrow$ P is positive definite matrix.
<br>
3. We can also see that since all terms in V are non-negative, $\lim_{\lVert w \rVert \to +\infty} V = +\infty$ so by definition V is radially unbounded.
<br>
4. We can show that $\dot{V}$ is negative definite:
\begin{align}
& \dot{V} = \omega_{1}(J_{1}*\dot\omega_{1}) + \omega_{2}(J_{2}*\dot\omega_{2}) + \omega_{3}(J_{3}*\dot\omega_{3}) \\
& = \omega_{1}(J_{2}-J_{3})\omega_{2}\omega_{3} - k_{1}\omega_{1}^{2} + \omega_{2}(J_{3}-J_{1})\omega_{1}\omega_{3} - k_{2}\omega_{2}^{2} + \omega_{3}(J_{1}-J_{2})\omega_{1}\omega_{1} - k_{3}\omega_{3}^{2} \\

& = 0 - k_{1}\omega_{1}^{2} - k_{2}\omega_{2}^{2} - k_{3}\omega_{3}^{2} \\
& = - (k_{1}\omega_{1}^{2} + k_{2}\omega_{2}^{2} + k_{3}\omega_{3}^{2}) < 0 \\

5. From 2., 3. 4., and by Lecture 3 pg 9, we prove that **the origin is GAS**
